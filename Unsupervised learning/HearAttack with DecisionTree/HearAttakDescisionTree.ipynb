```python
x = df1.drop('RiskOfCHD', axis=1)
y = df1['RiskOfCHD']
```

```python
df1['RiskOfCHD'].value_counts() 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() 
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from imblearn.over_sampling import SMOTE 

sm = SMOTE(random_state=42)

x_train_sample, y_train_sample = sm.fit_resample(x_train_scaled, y_train)

print(y_train.value_counts()) 
print(y_train_sample.value_counts())

```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42) 
model.fit(x_train_sample, y_train_sample)
```

```python
y_pred = model.predict(x_test_scaled) 
y_pred 
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15,8)) 
plot_tree(model, 
          filled=True,
          rounded=True,
          class_names=['No HD', "Yes HD"],
          feature_names=x.columns)

plt.show() 
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(
    model,
    x_test_scaled,
    y_test,
    display_labels=['No HD', 'Yes HD'],
    cmap=plt.cm.Blues
)
plt.show()
```

```python
path = model.cost_complexity_pruning_path(x_train_sample, y_train_sample)
ccp_alphas = path.ccp_alphas[:-1]
#ccp_alphas = ccp_alphas[:-1] 

model_dts = [] 

for ccp_alpha in ccp_alphas:
    model = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    model.fit(x_train_sample, y_train_sample) 
    model_dts.append(model)
```

```python
import warnings
warnings.filterwarnings('ignore')
train_scores = [model.score(x_train_sample, y_train_sample) for model in model_dts]
test_scores =[model.score(x_test, y_test) for model in model_dts]


fig, ax = plt.subplots() 
ax.set_xlabel("alpha") 
ax.set_ylabel("accuracy") 
ax.set_title("Accuracy vs alpha for training and testing sets") 
ax.plot(ccp_alphas, train_scores, marker='o', label='train', drawstyle='steps-post')
ax.plot(ccp_alphas, test_scores, marker='o', label='test', drawstyle='steps-post')
ax.legend() 
plt.show() 
```

```python
from sklearn.model_selection import cross_val_score
model = DecisionTreeClassifier(random_state=42, ccp_alpha=0.00001)


scores = cross_val_score(model, x_train_sample, y_train_sample, cv=5)
df = pd.DataFrame(data={'tree':range(5), 'accuracy':scores}) 

df.plot(x='tree', y='accuracy', marker='o', linestyle='--')
```

```python
alpha_loop_values = []

for ccp_alpha in ccp_alphas:
    model = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    scores = cross_val_score(model, x_train_sample, y_train_sample, cv=5)
    alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)]) 

alpha_results = pd.DataFrame(alpha_loop_values, columns=['alpha', 'mean_accuracy', 'std'])

alpha_results.plot(x='alpha',
                       y='mean_accuracy',
                       yerr='std',
                       marker='o',
                       linestyle='--')
```
